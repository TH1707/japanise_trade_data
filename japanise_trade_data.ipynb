```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('japanise_trade_data.csv')
print(df)
```

```python
print(df.head())
```

```python
print(df.tail())
```

```python
print(df.columns)
```

```python
print(df.isnull().sum)
```

```python
#Removining unnecessary columns
df_clean = df.drop(['Percent Change.1', 'Percent Change.2'], axis=1)
print(df_clean.head())
```

```python
print(df_clean.columns)
```

```python
print(df_clean.tail())
```

```python
#Basic statics for updated/cleaned data
df_clean.describe()
```

```python
# Visualizing trade balance among region or country
plt.figure(figsize=(10, 6))
sns.barplot(x="Exports", y="Balance", data=df_clean, palette='viridis')
plt.xticks(rotation=90)
plt.title('Exports vs Trade Balance by Country/Region')
plt.xlabel('Country/Region')
plt.ylabel('Trade Balance')
plt.show()
```

```python
#Visualize percent change in trade balance
plt.figure(figsize=(10, 6))
plt.figure(figsize=(12, 6))
sns.barplot(x='Exports', y='Percent Change', data=df_clean, palette='coolwarm')
plt.xticks(rotation=90)
plt.title('Exports vs Percent Change in Trade Balance')
plt.xlabel('Country/Region')
plt.ylabel('Percent Change')
plt.show()
```
